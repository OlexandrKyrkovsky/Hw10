```python
import pandas as pd
from surprise import SVD, SVDpp, NMF
from surprise import Dataset
from surprise.model_selection import cross_validate
```

```python
# Завантаження датасету MovieLens
data = Dataset.load_builtin('ml-100k')
```

```python
# Створення екземплярів для кожного алгоритму
svd = SVD()
svdpp = SVDpp()
nmf = NMF()
```

```python
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
cross_validate(svdpp, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
cross_validate(nmf, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

З даних результатів можна зробити висновок: MEAN_RMSE найменший у метода SVD++, але STD найбільший, проте все рівно дуже малий, тому можем зробити висновок що метод SVDpp на даному датасеті найкращий.
